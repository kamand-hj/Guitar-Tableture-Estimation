```python
def sox_it_out(hexpath):
    dir_name = hexpath.split(".")[0]
    os.mkdir(dir_name)
    output_mapping = {dir_name+'_0': {1: [1]},
                      dir_name+'_1': {1: [2]},
                      dir_name+'_2': {1: [3]},
                      dir_name+'_3': {1: [4]},
                      dir_name+'_4': {1: [5]},
                      dir_name+'_5': {1: [6]}
                      }
    for mix_type, remix_dict in output_mapping.items():
        tfm = sox.Transformer()
        tfm.remix(remix_dictionary=remix_dict)
        output_path = os.path.join(dir_name, '{}.wav'.format(mix_type))
        tfm.build(hexpath, output_path)
```

```python
def jam_it_in(filename, annotator):
    dir_name = filename.split(".")[0]
    jam = jams.JAMS()
    jam.file_metadata.duration = sox.file_info.duration(filename)
    jam.file_metadata.title = dir_name
    csv_files = glob.glob(dir_name+"/*.csv")
    csv_files.sort()
    for c in csv_files:
        with open(c, 'r') as stream:
            ann_csv = csv.reader(stream)
            ann_jam = jams.Annotation(namespace='pitch_midi', time=0,
                                      duration=jam.file_metadata.duration)
            ann_jam.annotation_metadata.data_source = os.path.basename(c.split('.')[0])
            ann_jam.annotation_metadata.annotator = annotator
            for row in ann_csv:
                try:
                    st = float(row[0])
                    midi_note = librosa.hz_to_midi(float(row[1]))[0]
                    dur = float(row[2])
                    ann_jam.append(time=st, value=midi_note, duration=dur,
                                   confidence=None)
                except:
                    print('empty row')
                    pass
        jam.annotations.append(ann_jam)
    out_path = dir_name + ".jams"
    jam.save(out_path) 
    return jam
```
